###### Area plot
<p> An area plot, also known as an area chart or an area graph, is a type of data visualization that displays quantitative data over a continuous interval or time period. It is similar to a line graph but with the area between the line and a baseline filled in with color or shading. </p>


###### Box Plot
<p>
A box plot, also known as a box-and-whisker plot, is a graphical representation of data that displays the distribution, central tendency, and variability of a dataset. It is often used to visualize the summary statistics of a dataset and identify potential outliers. </p>

###### Pie Chart
<p>A pie chart is a type of data visualization used to represent data as a circular graph. It is particularly useful for showing the composition or distribution of a whole when the data can be divided into categories or segments</p>
